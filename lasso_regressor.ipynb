```python
from sklearn.model_selection import train_test_split 

X = dm.data.iloc[:, 2:]
y = dm.data.iloc[:, 1]

X_train, X_test, y_train, y_test = train_test_split(X, y)

X_submission = dm.X_test_estimated[dm.X_test_estimated.columns.intersection(X.columns)]

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_submission.shape)

X
```

```python
import optuna
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LassoCV

def objective(trial):
    param = {
        'eps': trial.suggest_float('eps',1e-5, 1e-1),
        'n_alphas': trial.suggest_int('n_alphas', 20, 1000),
        'max_iter': trial.suggest_int('max_iter', 1, 10000),
        'tol': trial.suggest_float('tol', 1e-5, 1e-2),
        'cv': None
    }
    model = LassoCV(**param)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return mean_absolute_error(y_test, y_pred)

study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=50, timeout=600)
```
